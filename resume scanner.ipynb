```python
job_descripton=docx2txt.process('/content/accenture.docx')
resume=docx2txt.process('/content/e3be9e5b-71a5-4cd7-b571-a17d02ff9e14 (1).docx')
```

```python
print(resume)
```

```python
content=[job_descripton,resume]
```

```python
from sklearn.feature_extraction.text import CountVectorizer 
cv=CountVectorizer()
matrix= cv.fit_transform(content)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity_matrix = cosine_similarity(matrix)
```

```python
print(similarity_matrix)
```

```python
print('Resume matches by '+ str(similarity_matrix[1][0]*100)+'%')
```
